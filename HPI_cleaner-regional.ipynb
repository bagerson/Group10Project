```python
#import data from CSV
HPI_df = pd.read_csv(file_four)
HPI_df.head()

```

```python
#drop unnecessary columns
HPI_working = HPI_df.drop(columns = ['hpi_type', 'hpi_flavor', 'level', 'place_id', 'index_sa'])
HPI_working.head()
```

```python
HPI_working.size
```

```python
#remove extra place names, only want regional and U.S.
#split by geographic region
# the data is divided into 9 census divisions (+puerto rico?)

place_names = HPI_working["place_name"].unique()
#print(place_names)

#Note: the HPI DATA contains a huge number of place names.  Want only regional and U.S.
```

```python
#shorten the place names list to the first 10 elements.  9 region + U.S. average
place_names = place_names[:10]
place_names
```

```python
#remove quarterly data

HPI_working = HPI_working.drop(HPI_working[HPI_working.frequency == "quarterly"].index)

HPI_working.head()
```

```python
HPI_working.size
```

```python
#remove unnecessary rows so data is 01/2009+

HPI_working = HPI_working.loc[HPI_working['yr'] >= 2009]
HPI_working.head()
```

```python
HPI_working.size
```

## NOTE: DATA starting in 2009 must be "zeroed" because is it a referant of 1991 data as %

```python
#extract U.S. placename data
US_HPI = HPI_working.loc[HPI_working['place_name'] == place_names[9]]
US_HPI.head()
```

```python
US_HPI.size
```

```python
#create dataframes for each region
ENC_HPI = HPI_working.loc[HPI_working['place_name'] == place_names[0]]
ESC_HPI = HPI_working.loc[HPI_working['place_name'] == place_names[1]]
MAD_HPI = HPI_working.loc[HPI_working['place_name'] == place_names[2]]
MD_HPI = HPI_working.loc[HPI_working['place_name'] == place_names[3]]
NED_HPI = HPI_working.loc[HPI_working['place_name'] == place_names[4]]
PD_HPI = HPI_working.loc[HPI_working['place_name'] == place_names[5]]
SAD_HPI = HPI_working.loc[HPI_working['place_name'] == place_names[6]]
WNC_HPI = HPI_working.loc[HPI_working['place_name'] == place_names[7]]
WSC_HPI = HPI_working.loc[HPI_working['place_name'] == place_names[8]]
```
